```python
# Confidence Interval for the Mean
def confidence_interval(data, confidence=0.95):
    mean = data.mean()
    std = data.std()
    n = len(data)
    z = stats.norm.ppf(1 - (1 - confidence) / 2)  # z-score for 95% CI
    margin_of_error = z * (std / np.sqrt(n))
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error
    return (lower_bound, upper_bound)

ci_results = {}
for col in at54_non_cat_vars:
    ci_results[col] = confidence_interval(at54_sample[col])


print("\nConfidence Intervals (95%):")
for col, ci in ci_results.items():
    print(f"{col}: {ci[0]:.2f} to {ci[1]:.2f}")
```

#### **> Observations:**

**Precise Mean Estimation:** Narrow CI ranges for Quantity, Value, and Weight indicate precise estimation of their population means. We can be 95% confident that the true population means lie within these intervals.

**Centered Around 0.5:** The CIs are centered around 0.5, suggesting that the distributions are likely centered around this value after MinMax scaling.

**Slightly Higher Variability for Weight and Value:** The slightly wider CI for Weight and Value and Quantity (0.49 to 0.51) indicates a marginally higher uncertainty in estimating its population mean.

## **Inferential Statistics**

```python
# Test of Homogeneity {Chi-sq} on categorical data

at54_contingency_table = pd.crosstab(at54_sample['Import_Export'], at54_sample['Category'])

# Perform the Chi-square test
chi2, p, dof, expected = chi2_contingency(at54_contingency_table)

print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of freedom: {dof}")
print("Expected frequencies:")
print(expected)

# Interpret the results
at54_alpha = 0.05  # Significance level

if p < at54_alpha:
    print("Reject the null hypothesis: There is a significant association between the two categorical variables.")
else:
    print("There is no significant association between the two categorical variables.")
```

```python
# Test of Normality {Shapiro-Wilk | Kolmogorov-Smirnov | Anderson-Darling | Jarque-Bera

def normality_tests(data):

    at54_results = {}

    # Shapiro-Wilk test
    shapiro_statistic, shapiro_pvalue = shapiro(data)
    at54_results["Shapiro-Wilk"] = {"statistic": shapiro_statistic, "p-value": shapiro_pvalue}

    # Kolmogorov-Smirnov test (against a normal distribution)
    ks_statistic, ks_pvalue = kstest(data, 'norm')
    at54_results["Kolmogorov-Smirnov"] = {"statistic": ks_statistic, "p-value": ks_pvalue}

    # Anderson-Darling test
    anderson_result = anderson(data)
    at54_results["Anderson-Darling"] = {
        "statistic": anderson_result.statistic,
        "critical_values": anderson_result.critical_values,
        "significance_level": anderson_result.significance_level
    }

    # Jarque-Bera test
    jb_statistic, jb_pvalue = jarque_bera(data)
    at54_results["Jarque-Bera"] = {"statistic": jb_statistic, "p-value": jb_pvalue}

    return at54_results
```

```python
from scipy import stats

# Test of Correlation (T-test)
def t_test_correlation(data1, data2):
    t_statistic, p_value = stats.ttest_ind(data1, data2)
    return {"t-statistic": t_statistic, "p-value": p_value}

for col in at54_non_cat_vars:
    print(f"Normality tests for {col}:")
    normality_test_results = normality_tests(at54_sample[col])
    for test_name, result in normality_test_results.items():
        print(f"  {test_name}: {result}")

    # Example of T-test
    if len(at54_non_cat_vars) > 1:
        for i in range(len(at54_non_cat_vars)):
            for j in range(i+1, len(at54_non_cat_vars)):
                col1, col2 = at54_non_cat_vars[i], at54_non_cat_vars[j]
                t_test_result = t_test_correlation(at54_sample[col1], at54_sample[col2])
                print(f"T-test Correlation between {col1} and {col2}: {t_test_result}")
    print()
    print('*'*20)
    print()
```

# Model Run Statistics

## Time Taken

```python
import time
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

def time_model_run(model, X_train, y_train):
    start_time = time.time()
    model.fit(X_train, y_train)
    end_time = time.time()
    return end_time - start_time

# Logistic Regression
log_reg_model = LogisticRegression(random_state=55003)
log_reg_time = time_model_run(log_reg_model, at54_X_train, at54_y_train)
print(f"Logistic Regression runtime: {log_reg_time:.4f} seconds")

# Support Vector Machine (SVM)
svm_model = SVC(random_state=55003)
svm_time = time_model_run(svm_model, at54_X_train, at54_y_train)
print(f"SVM runtime: {svm_time:.4f} seconds")

# Stochastic Gradient Descent (SGD)
sgd_model = SGDClassifier(random_state=55003)
sgd_time = time_model_run(sgd_model, at54_X_train, at54_y_train)
print(f"SGD runtime: {sgd_time:.4f} seconds")

# Decision Tree
dt_model = DecisionTreeClassifier(random_state=55003)
dt_time = time_model_run(dt_model, at54_X_train, at54_y_train)
print(f"Decision Tree runtime: {dt_time:.4f} seconds")

# K-Nearest Neighbors (KNN)
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_time = time_model_run(knn_model, at54_X_train, at54_y_train)
print(f"KNN runtime: {knn_time:.4f} seconds")

# Naive Bayes
nb_model = GaussianNB()
nb_time = time_model_run(nb_model, at54_X_train, at54_y_train)
print(f"Naive Bayes runtime: {nb_time:.4f} seconds")

# Random Forest
rf_model = RandomForestClassifier(random_state=55003)
rf_time = time_model_run(rf_model, at54_X_train, at54_y_train)
print(f"Random Forest runtime: {rf_time:.4f} seconds")

# Extreme Gradient Boosting (XGBoost)
xgb_model = XGBClassifier(random_state=55003)
xgb_time = time_model_run(xgb_model, at54_X_train, at54_y_train)
print(f"XGBoost runtime: {xgb_time:.4f} seconds")
```

### Observations:

- **Fastest:** KNN and Naive Bayes are the fastest models, taking less than 0.01 seconds.
- **Moderate:** Logistic Regression, SGD, and Decision Tree have moderate runtimes, ranging from 0.0155 to 0.0534 seconds.
- **Relatively Slow:** XGBoost has a relatively slow runtime, taking 0.5870 seconds.
- **Slowest:** SVM and Random Forest are the slowest models, taking 0.9777 seconds and 1.5991 seconds, respectively.

### Insights:

1. **KNN and Naive Bayes:** These models are ideal for quick analysis or when computational resources are limited.
2. **Logistic Regression, SGD, and Decision Tree:** These models strike a good balance between speed and complexity, making them suitable for many applications that require moderately fast results with reasonable accuracy.
3. **XGBoost:** With a moderate runtime, XGBoost offers a balance between speed and accuracy, making it a good choice for more complex datasets.
4. **SVM and Random Forest:** These models have the slowest runtimes and require more computational resources. They are better suited for tasks where accuracy is critical and sufficient computational resources are available.

### Recommendations:

- **For time-sensitive tasks:** Consider using KNN, Naive Bayes, or Logistic Regression.
- **For balanced performance:** Use Decision Tree, SGD, or XGBoost for their trade-off between runtime and accuracy.
- **For accuracy-critical tasks:** Explore Random Forest or SVM if runtime is not a constraint and accuracy is a priority.



## Memory Utilized

```python
import tracemalloc
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

def get_model_memory_usage(model, X, y):
    tracemalloc.start()
    model.fit(X, y)
    current, peak = tracemalloc.get_traced_memory()
    tracemalloc.stop()
    return peak / (1024 * 1024)  # Convert bytes to MB

# Logistic Regression
log_reg_model = LogisticRegression(random_state=55003)
log_reg_memory = get_model_memory_usage(log_reg_model, at54_X_train, at54_y_train)
print(f"Logistic Regression memory usage: {log_reg_memory:.2f} MB")

# Support Vector Machine (SVM)
svm_model = SVC(random_state=55003)
svm_memory = get_model_memory_usage(svm_model, at54_X_train, at54_y_train)
print(f"SVM memory usage: {svm_memory:.2f} MB")

# Stochastic Gradient Descent (SGD)
sgd_model = SGDClassifier(random_state=55003)
sgd_memory = get_model_memory_usage(sgd_model, at54_X_train, at54_y_train)
print(f"SGD memory usage: {sgd_memory:.2f} MB")

# Decision Tree
dt_model = DecisionTreeClassifier(random_state=55003)
dt_memory = get_model_memory_usage(dt_model, at54_X_train, at54_y_train)
print(f"Decision Tree memory usage: {dt_memory:.2f} MB")

# K-Nearest Neighbors (KNN)
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_memory = get_model_memory_usage(knn_model, at54_X_train, at54_y_train)
print(f"KNN memory usage: {knn_memory:.2f} MB")

# Naive Bayes
nb_model = GaussianNB()
nb_memory = get_model_memory_usage(nb_model, at54_X_train, at54_y_train)
print(f"Naive Bayes memory usage: {nb_memory:.2f} MB")

# Random Forest
rf_model = RandomForestClassifier(random_state=55003)
rf_memory = get_model_memory_usage(rf_model, at54_X_train, at54_y_train)
print(f"Random Forest memory usage: {rf_memory:.2f} MB")

# Extreme Gradient Boosting (XGBoost)
xgb_model = XGBClassifier(random_state=55003)
xgb_memory = get_model_memory_usage(xgb_model, at54_X_train, at54_y_train)
print(f"XGBoost memory usage: {xgb_memory:.2f} MB")
```

### Observations:

- **Highest:** XGBoost has the highest memory usage, requiring 0.85 MB.
- **Lowest:** Naive Bayes has the lowest memory usage, requiring only 0.11 MB.
- **Moderate:** Logistic Regression, SVM, Decision Tree, and KNN have moderate memory usage, ranging from 0.22 MB to 0.34 MB.
- **Relatively High:** Random Forest has relatively high memory usage, requiring 0.61 MB.
- **Low:** SGD has lower memory usage, requiring only 0.18 MB.

### Insights:

1. **XGBoost:** With the highest memory usage, it requires careful resource management and is best suited for tasks with ample memory resources.
2. **Random Forest:** Has relatively high memory usage but provides strong accuracy for complex tasks, making it suitable when memory resources are moderately available.
3. **Logistic Regression, SVM, Decision Tree, and KNN:** These models have moderate memory requirements, making them suitable for most use cases without significant resource constraints.
4. **SGD:** With low memory usage, it is ideal for resource-constrained environments requiring fast computation.
5. **Naive Bayes:** With the lowest memory footprint, it is excellent for memory-sensitive tasks or when running multiple models in parallel.

### Recommendations:

- **For memory-sensitive tasks:** Use Naive Bayes or SGD due to their minimal memory requirements.
- **For balanced memory usage and performance:** Consider Logistic Regression, SVM, Decision Tree, or KNN.
- **For accuracy-critical tasks with ample memory resources:** Explore Random Forest or XGBoost.
